```python
import pandas as pd

# Load data
data=pd.read_csv('commits.csv')
data_issues=pd.read_csv('issues.csv')

# If 'key' is equal append on right, else append on bottom-right
data = data.join(data_issues.set_index('key'), on='key', how = 'outer')
data = data.reset_index(drop=True) # Reset index to start from 0 to n_samples - 1
```

```python
# Fill empry fields
float_cols = data.select_dtypes(include=['float64']).columns
str_cols = data.select_dtypes(include=['object']).columns

data.loc[:, float_cols] = data.loc[:, float_cols].fillna(0)
data.loc[:, str_cols] = data.loc[:, str_cols].fillna('')

# fix bug in datetime format in csv files 
# eg.datetime was 0010-10-28T10:07:00.000+0000 need to be 2010-10-28T10:07:00.000+000
for i in data.index:
    if data.loc[i, 'created'][:2]=='00':
        data.loc[i, 'created']='20' + data.loc[i, 'created'][2:]
```

```python
# sort by column 'created'
data.sort_values(by=['created'])
```

```python
# group by 'author name' and top 10 most productive
data.groupby(['author_name']).count().sort_values(['key'], ascending=False).head(10)
```

```python
# filter issuetype == 'Bug'
data[data['issuetype']=='Bug']
```

```python
# print commits between 1 April 2016 and 1 October 2016 and severity==10
data[(data['created']>'2016-04-01')&(data['created']<'2016-10-01')&(data['severity']==10)]
```

```python
# count commits between 1 April 2016 and 1 October 2016 and severity==10
data[(data['created']>'2016-04-01')&(data['created']<'2016-10-01')&(data['severity']==10)].shape[0]
```

```python
# top 10 most longer commits
data['duration']=pd.to_datetime(data['resolved'])-pd.to_datetime(data['created'])
data.sort_values(by=['duration'], ascending=False).head(10)
```
